<h3>total_bill VS tip:</h3>

primero, realizamos la prueba levene para determinar si las muestras tienen varianzas homogeneas y de este modo determinar que coeficiente de correlación determinar:

```python
stat, pvalor = stats.levene(tips['total_bill'], tips['tip'])
if pvalor < 0.05:  
  print(f'Las muestras no tienen varianzas homogéneas (rechazar H0) p = {pvalor:.3f}')
else:
  print(f'Las muestras tienen varianza homogéneas (no rechazar H0) p = {pvalor:.3f}')
```

```python
def correlacion_spearmanr_con_significancia(a, b):
    coef, p = stats.spearmanr(a, b)
    print(f'Coeficiente de correlación de Spearman: {coef:.2f}')

    if p > 0.05:
        print(f'Las muestras no están correlacionadas (no rechazar H0) p = {p:.3f}')
    else:
        print(f'Las muestras están correlacionadas (rechazar H0) p = {p:.3f}')
    return coef
```

Se determina que se debe hallar el coeficiente de correlación $\rho$ de Spearman.

```python
r=correlacion_spearmanr_con_significancia(tips['total_bill'], tips['tip'])
```

Segundo, realizamos una visualización en la gráfica de correlación para determinar posibles valores atipicos.

```python

tips.plot.scatter('total_bill','tip',title=f'Correlación de total_bill y tip: {r:.2}')
```

Se evidencia que se pueden determinar ciertos valores atipicos que disminuyen la correlación de las variables. Es por esto que vamos a realizar una limpieza de los datos solo con cumplir los fines academicos de este proyecto.

```python
outliers = tips[(tips['total_bill'] >= 30) & (tips['tip'] <5) | (tips['total_bill'] <= 10) & (tips['tip'] >3.5) ]
outliers
ax=tips.plot.scatter('total_bill','tip' ,title=f'Correlación de total_bill y tip: {r:.2}',
                     s = 70, alpha = 0.6, color = '#35bacc')
outliers.plot.scatter('total_bill', 'tip' ,s = 70, alpha = 0.8,color = '#f96340',ax=ax);
```

Se acotan los valores atípicos que queremos descartar y posteriormente, realizaremos el filtrado de los datos de interés.
Es de vital importancia resaltar que este procedimiento solo se hace con fines academicos, y no se modificara el dataframe principal. Pues a opinión del autor, no es necesario descartar estos datos especificos.

```python
# Construimos una máscara para filtrar los valores atípicos
mask = tips.isin(outliers)
tips_limpio = tips.mask(mask).dropna() 
```

```python
r = correlacion_spearmanr_con_significancia(tips_limpio['total_bill'], tips_limpio['tip'])
print(r)
```

Observe que el coeficiente se incrementó considerablemente ($r = 0.75$). Veamos el nuevo diagrama de dispersión entre las variables:

```python
tips_limpio.plot.scatter('total_bill', 'tip', title= f'Correlación sin outliers: {r:.2}',
                     s = 70, alpha = 0.6, color = '#35bacc', figsize = (7,5));
```

#### Regresión lineal:


con la libreria Statsmodel.formula.api

```python
# La fórmula se pone de la forma: VARIABLE_DEPENDIENTE ~ VARIABLE_INDEPENDIENTE 
model = smf.ols("total_bill ~ tip", data=tips_limpio).fit() # Aquí creamos el modelo y calculamos sus parámetros directamente con fit()
model.summary()
```
