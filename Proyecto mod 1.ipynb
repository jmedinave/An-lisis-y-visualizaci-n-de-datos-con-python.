<h3>total_bill VS tip:</h3>

primero, realizamos la prueba levene para determinar si las muestras tienen varianzas homogeneas y de este modo determinar que coeficiente de correlaci√≥n determinar:

```python
stat, pvalor = stats.levene(tips['total_bill'], tips['tip'])
if pvalor < 0.05:  
  print(f'Las muestras no tienen varianzas homog√©neas (rechazar H0) p = {pvalor:.3f}')
else:
  print(f'Las muestras tienen varianza homog√©neas (no rechazar H0) p = {pvalor:.3f}')
```

```python
def correlacion_spearmanr_con_significancia(a, b):
    coef, p = stats.spearmanr(a, b)
    print(f'Coeficiente de correlaci√≥n de Spearman: {coef:.2f}')

    if p > 0.05:
        print(f'Las muestras no est√°n correlacionadas (no rechazar H0) p = {p:.3f}')
    else:
        print(f'Las muestras est√°n correlacionadas (rechazar H0) p = {p:.3f}')
    return coef
```

Se determina que se debe hallar el coeficiente de correlaci√≥n $\rho$ de Spearman.

```python
r=correlacion_spearmanr_con_significancia(tips['total_bill'], tips['tip'])
```

Segundo, realizamos una visualizaci√≥n en la gr√°fica de correlaci√≥n para determinar posibles valores atipicos.

```python

tips.plot.scatter('total_bill','tip',title=f'Correlaci√≥n de total_bill y tip: {r:.2}')
```

Se evidencia que se pueden determinar ciertos valores atipicos que disminuyen la correlaci√≥n de las variables. Es por esto que vamos a realizar una limpieza de los datos solo con cumplir los fines academicos de este proyecto.

```python
outliers = tips[(tips['total_bill'] >= 30) & (tips['tip'] <5) | (tips['total_bill'] <= 10) & (tips['tip'] >3.5) ]
outliers
ax=tips.plot.scatter('total_bill','tip' ,title=f'Correlaci√≥n de total_bill y tip: {r:.2}',
                     s = 70, alpha = 0.6, color = '#35bacc')
outliers.plot.scatter('total_bill', 'tip' ,s = 70, alpha = 0.8,color = '#f96340',ax=ax);
```

Se acotan los valores at√≠picos que queremos descartar y posteriormente, realizaremos el filtrado de los datos de inter√©s.
Es de vital importancia resaltar que este procedimiento solo se hace con fines academicos, y no se modificara el dataframe principal. Pues a opini√≥n del autor, no es necesario descartar estos datos especificos.

```python
# Construimos una m√°scara para filtrar los valores at√≠picos
mask = tips.isin(outliers)
tips_limpio = tips.mask(mask).dropna() 
```

```python
r = correlacion_spearmanr_con_significancia(tips_limpio['total_bill'], tips_limpio['tip'])
print(r)
```

Observe que el coeficiente se increment√≥ considerablemente ($r = 0.75$). Veamos el nuevo diagrama de dispersi√≥n entre las variables:

```python
tips_limpio.plot.scatter('total_bill', 'tip', title= f'Correlaci√≥n sin outliers: {r:.2}',
                     s = 70, alpha = 0.6, color = '#35bacc', figsize = (7,5));
```

#### Regresi√≥n lineal:


con la libreria Statsmodel.formula.api

```python
# La f√≥rmula se pone de la forma: VARIABLE_DEPENDIENTE ~ VARIABLE_INDEPENDIENTE 
model = smf.ols("total_bill ~ tip", data=tips_limpio).fit() # Aqu√≠ creamos el modelo y calculamos sus par√°metros directamente con fit()
model.summary()
```
